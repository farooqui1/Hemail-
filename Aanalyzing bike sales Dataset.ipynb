## Analyzing the Database

### Which brand has the highest listing price

```python
SELECT b.brand_name, AVG(p.list_price) as avg_list_price
FROM production.brands as b
INNER JOIN production.products as p
ON b.brand_id = p.brand_id
GROUP BY b.brand_name
ORDER BY avg_list_price DESC
```

### How much the prices changed over the years of electric bikes

```python
SELECT p.model_year, AVG(p.list_price) as avg_list_price
FROM production.products as p
INNER JOIN production.categories as c
ON p.category_id = c.category_id
WHERE c.category_name = 'Electric Bikes'
GROUP BY p.model_year
ORDER BY p.model_year
```

### Which brand has the highest number of stocks

```python
SELECT b.brand_name, SUM(s.quantity) as total_stocks
FROM production.brands as b
INNER JOIN production.products as p
ON b.brand_id = p.brand_id
INNER JOIN production.stocks as s
ON p.product_id = s.product_id
GROUP BY b.brand_name
ORDER BY total_stocks DESC
```

### Bikes of which category are the most expensive

```python
SELECT c.category_name, ROUND(AVG(p.list_price),2) as avg_list_price
FROM production.categories as c
INNER JOIN production.products as p
ON c.category_id = p.category_id
GROUP BY c.category_name
ORDER BY avg_list_price DESC
```

### How many stores are there with respect to cities

```python
SELECT city,COUNT(store_name) as num_stores
FROM sales.stores
GROUP BY city
ORDER BY num_stores
```

### How many quantities of bikes are in stock in each store

```python
SELECT s.store_name, SUM(ot.quantity) as total_quantity
FROM sales.stores as s
INNER JOIN sales.orders as o
ON s.store_id = o.store_id
INNER JOIN sales.order_items as ot
ON o.order_id = ot.order_id
GROUP BY s.store_name
ORDER BY total_quantity DESC
```

### Which brand has the hoghest orders

```python
SELECT b.brand_name, COUNT(DISTINCT o.order_id) AS num_items
FROM production.brands as b
INNER JOIN production.products as p
ON b.brand_id = p.brand_id
INNER JOIN sales.order_items as o
ON p.product_id = o.product_id
GROUP BY b.brand_name
ORDER BY num_items DESC
```
