### No of employees in each department

```python
SELECT d.dept_name,COUNT(e.emp_no) as num_emp
FROM departments as d
INNER JOIN dept_emp as e
ON d.dept_no = e.dept_no
GROUP BY d.dept_name
ORDER BY num_emp DESC
```

#### Development has the highest number of employees

### Now, let's see the number of employees with respect ot their titles

```python
SELECT t.title,COUNT(e.emp_no) as num_emp
FROM titles as t
INNER JOIN dept_emp as e
ON t.emp_no = e.emp_no
GROUP BY t.title
ORDER BY num_emp DESC
```

#### Engineers are the most in the company

### Let's see the min and max date of employees

```python
SELECT emp_no,
       MIN(from_date) as start_date,
       MAX(to_date)as end_date
FROM salaries
```

#### We can se see that we have an error in our dataset in the year of max date so we will not use this value in our analysis

```python
SELECT emp_no,
       MIN(from_date) as start_date,
       MAX(to_date)as end_date
FROM salaries
WHERE to_date != (SELECT MAX(to_date) FROM salaries)
GROUP BY emp_no
LIMIT 10
```

#### Seems fine...

### For the duration of each employee

```python
WITH employees_date AS(
SELECT emp_no,
       MIN(from_date) as start_date,
       MAX(to_date)as end_date
FROM salaries
WHERE to_date != (SELECT MAX(to_date) FROM salaries)
GROUP BY emp_no
)

SELECT emp_no,start_date,
       end_date,
       DATEDIFF(end_date,start_date) / 365 as emp_tenure
FROM employees_date
GROUP BY emp_no
LIMIT 10
```

### For average employee tenure respect to title

```python
WITH employees_date AS(
SELECT emp_no,
       MIN(from_date) as start_date,
       MAX(to_date)as end_date
FROM salaries
WHERE to_date != (SELECT MAX(to_date) FROM salaries)
GROUP BY emp_no
),
employee_tenure as(
SELECT emp_no,start_date,
       end_date,
       DATEDIFF(end_date,start_date) / 365 as emp_tenure
FROM employees_date
GROUP BY emp_no
)
SELECT t.title,AVG(e.emp_tenure) as avg_emp_tenure
from employee_tenure as e
INNER JOIN titles as t
ON e.emp_no = t.emp_no
AND e.end_date = t.to_date
GROUP BY t.title
```

### For minimum and maximum salas of employees

```python
SELECT MIN(salary) as min_salary,
       MAX(salary)as max_salary
FROM salaries
WHERE to_date != (SELECT MAX(to_date) FROM salaries)
```

### For salary ranges of employees

```python
SELECT  t.title,
SUM(CASE WHEN s.salary >40000 THEN 1 ELSE 0 END) as avg_salary,
SUM(CASE WHEN s.salary >70000 THEN 1 ELSE 0 END) as better_salary,
SUM(CASE WHEN s.salary >100000 THEN 1 ELSE 0 END) as good_salary
FROM titles as t
INNER JOIN salaries as s
ON t.emp_no = s.emp_no
INNER JOIN employees as e
ON t.emp_no = e.emp_no
GROUP BY t.title
```
